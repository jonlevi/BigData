## Step 4.3

```python
# TODO: Create dataframe of airline_iata, number of incidents, number of routes
num_incidents_df = incidents_df.groupby(by='Airline').count().reset_index() \
            .sort_values('FlightNum', ascending=False)[['Airline', 'FlightNum']] \
            .rename(index=str, columns={'FlightNum':'Number of Incidents', 'Airline':'name'})
num_incidents_df = num_incidents_df[num_incidents_df['Number of Incidents']>2]

#EgyptAir vs Eggyptair bug
num_incidents_df['name'] = num_incidents_df['name'].str.lower()

num_routes_df = routes_df.groupby(by = 'airline_id').size().to_frame().reset_index()
num_routes_df.columns = ['airline_id', 'Number of Routes']

#EgyptAir vs Eggyptair bug
num_incidents_df['name'] = num_incidents_df['name'].str.lower()
airlines_df['name'] = airlines_df['name'].str.lower()

incidents_and_names_df = pd.merge(airlines_df[['airline_id', 'name', 'IATA']], num_incidents_df,
                  on='name')
incidents_and_routes_df = pd.merge(incidents_and_names_df[['airline_id', 'IATA', 'Number of Incidents']], 
                                   num_routes_df, on='airline_id').drop(['airline_id'], axis=1)
incidents_and_routes_df
                                   
```

```python
# TODO: Scatter plot of number of routes vs. number of incidents
plt.figure()
plt.scatter(incidents_and_routes_df['Number of Routes'], incidents_and_routes_df['Number of Incidents'], marker = 'o')
for index,row in incidents_and_routes_df.iterrows():
    x = int(row['Number of Routes'])
    y = int(row['Number of Incidents'])
    plt.annotate(row['IATA'], xy=(x,y))
plt.title('Number of Routes vs. Number of Incidents');
```

```python
# TODO: get random sample of flights from SQL
flights_s_df = pd.read_sql_query('SELECT * FROM flights_db ORDER BY RANDOM() LIMIT 1000', con=engine)
flights_s_df
```

```python
# TODO: create nested map of Jaccard distance

#takes in two sets and computes jaccard distance
def jaccard(A, B):
    AinterB = A.intersection(B)
    AunionB = A.union(B)
    return len(AinterB)/len(AunionB)

r_to_p_to_d = {}
for c1 in routes_df:
    if routes_df[c1].dtype is not pd.np.dtype(object):
        continue
        
    r_to_p_to_d[c1] = {}
    A = set(routes_df[c1])
    for c2 in flights_s_df:
        B = set(flights_s_df[c2])
        j = jaccard(A,B)
        if (j>0):
            r_to_p_to_d[c1][c2] = j
        
```

## Step 5.2.1

```python
# TODO: create joint DataFrame flights_s_df 
flights_s_df = routes_df.merge(flights_s_df, left_on = ['airline', 'dest_airport', 'source_airport'],
                               right_on = ['Carrier', 'Dest', 'Origin'])
flights_s_df
```

```python
# TODO: output for each airline IATA code and destination city IATA code, 
#the number of flights that appear in the sample flights_s_df
num_flights_per_IATA_df = flights_s_df.groupby(by=['airline', 'dest_airport']).size().reset_index() \
            .rename(index=str, 
            columns={'airline':'airline_IATA', 
                    'dest_airport':'dest_airport_IATA',
                    0: 'Num Flights in Sample'})
num_flights_per_IATA_df.set_index(['airline_IATA', 'dest_airport_IATA'])
```

## Step 5.2.2

```python
# TODO: output bad flights
bad_flights_s_df = flights_s_df[(flights_s_df['Cancelled']==1.0) | (flights_s_df['ArrDelayMinutes'] >= 30)]
num_bad_flights_per_IATA_df = bad_flights_s_df.groupby(by=['airline', 'dest_airport']).size().reset_index() \
            .rename(index=str, 
            columns={'airline':'airline_IATA', 
                    'dest_airport':'dest_airport_IATA',
                    0: 'Num Bad Flights in Sample'})
num_bad_flights_per_IATA_df.set_index(['airline_IATA', 'dest_airport_IATA'])
```

## Step 5.2.3

```python
# TODO: pairs with more than one bad flight
multiple_bad_flights_per_IATA = num_bad_flights_per_IATA_df \
                [num_bad_flights_per_IATA_df['Num Bad Flights in Sample']>1].reset_index().drop('index', axis=1)
multiple_bad_flights_per_IATA.set_index(['airline_IATA', 'dest_airport_IATA'])
```

```python
# TODO: bar chart for bad flights
plt.figure()
numRows = multiple_bad_flights_per_IATA.shape[0]
labels = [(row['airline_IATA'] + "\n" + row['dest_airport_IATA']) \
          for index,row in multiple_bad_flights_per_IATA.iterrows()]
plt.bar(np.arange(0,numRows), multiple_bad_flights_per_IATA['Num Bad Flights in Sample'],
                       tick_label = labels)
plt.ylabel('Number of Bad Flights in Sample')
plt.title('Number of Delayed/Cancelled Flights per Airline-Destination Pair');
```
