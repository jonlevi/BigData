```python
# fillna

# Replace NaNs with blanks if the column is a string, so we donâ€™t get
# objects in the column
def fillna_col(series):
    if series.dtype is pd.np.dtype(object):
        return series.fillna('')
    else:
        return series
```

```python
# TODO: Define nullify here

# Replace \Ns with NaNs
def nullify(x):
    return np.NaN if x=='\\N' else x
```

```python
# TODO: regularize and remove nulls according to Step 2
airlines_df = airlines_df.applymap(nullify)
airports_df = airports_df.applymap(nullify)
routes_df = routes_df.applymap(nullify)
flights_df = flights_df.applymap(nullify)
```

```python
routes_df = routes_df.dropna(subset=['airline_id', 'source_airport_id', 'dest_airport_id'], how='any')
```

```python
airlines_df = airlines_df.apply(fillna_col)
airports_df = airports_df.apply(fillna_col)
routes_df = routes_df.apply(fillna_col)
flights_df = flights_df.apply(fillna_col)
```

```python
labels = ['airline_id', 'source_airport_id', 'dest_airport_id']
for s in labels:
    routes_df[s] = routes_df[s].astype(int)
```

```python
# TODO: read and clean raw text file and store in clean_incidents
clean_incidents = []
edit_regex = re.compile("\[edit\]")
current_year = ""
current_str = ""
modern = False
count = 0

with open('incidents_raw.txt') as file:
    for line in file:
        count += 1
        line_edit = re.sub(edit_regex, '', line)
        if (line_edit.strip().isdigit()):

            if (int(line_edit)<1997): 
                modern = False
                continue
            else: modern = True
                
            current_year = line_edit
            
             
        elif ((' â€“ ' in line_edit) and modern and ('Flight' in line_edit)):
            to_add = current_year + line_edit.replace(" â€“ ", " , ")
            to_add = to_add.replace("\n", " ")
            clean_incidents.append(to_add)
clean_incidents
```

```python
def get_airline(messyString):
    airline_regex = re.compile('([A-Z]\w+(\s[A-Z][a-z]+)*) Flight')
    return re.findall(airline_regex, messyString)[0][0]

def get_flight_number(messyString):
    flight_num_regex = re.compile('Flight\s[0-9]*\w*')
    return re.findall(flight_num_regex, messyString)[0].replace('Flight ', '')
```

```python
# TODO: convert clean_incidents into dataframe incidents_df
incidents_df = pd.DataFrame(columns=['Date', 'Airline', 'FlightNum'])
airline_regex = re.compile('â€“ .*Flight')

for incident in clean_incidents:
    date = incident.split(' , ')[0]
    airline = get_airline(incident)
    flight_num = get_flight_number(incident)
    incidents_df = incidents_df.append({'Date':date, 'Airline':airline, 'FlightNum':flight_num}, ignore_index=True)
    
```

```python
import datetime
# TODO: change column type of 'Date'
incidents_df['Date'] = pd.to_datetime(incidents_df['Date'])
incidents_df = incidents_df.drop_duplicates()
```
